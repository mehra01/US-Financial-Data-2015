```python
log_clf = LogisticRegression()
rnd_clf = RandomForestClassifier(random_state=42, max_leaf_nodes=20, max_depth=7)
svm_clf = SVC(kernel='poly', gamma='auto', coef0=1, shrinking=False, probability=True)
xgb_clf = XGBClassifier(n_jobs = -1, random_state=42, n_estimators = 9,
                     learning_rate=0.4, max_depth = 3, subsample=0.9,
                     colsample_bytree = 0.1)
voting_clf = VotingClassifier(
    estimators=[('lr', log_clf), ('rf', rnd_clf), ('xgb', xgb_clf)],
    voting='soft')

# ('lr', log_clf), ('rf', rnd_clf), ('svc', svm_clf), ('xgb', xgb_clf)
```

```python
def predict_scores(train_inputs, train_targets, val_inputs, val_targets, model= voting_clf):
    model.fit(train_inputs, train_targets)
    train_preds = model.predict(train_inputs)
    val_preds = model.predict(val_inputs)
    train_score = accuracy_score(train_preds, train_targets)
    val_score = accuracy_score(val_preds, val_targets)
    return train_score, val_score
```

```python
train_score, val_score = predict_scores(train2015_inputs, train2015_targets, val2015_inputs, val2015_targets, voting_clf)
train_score, val_score
```
